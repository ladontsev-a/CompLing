```python
import re
```

```python
def remove_tags(text):
    clean_text = re.sub('<.+?>', ' ', text)
    clean_text = re.sub(' +', ' ', clean_text)
    clean_text = re.sub(' \.', '.', clean_text)
    clean_text = re.sub(' ,', ',', clean_text)
    clean_text = re.sub(' \n', '\n', clean_text)
    return clean_text
clean_text = remove_tags(text)
clean_text
```

## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
```

```python
mystem = Mystem()
morph = MorphAnalyzer()
analyzed = mystem.analyze(clean_text)
analyzed[:10]
```

```python
stem_tokens = [lex['text'] for lex in analyzed if lex.get('analysis')]
stem_tokens[:10]
```

```python
stem_lem = [lex['analysis'][0]['lex'] for lex in analyzed if lex.get('analysis')]
stem_lem[:10]
```

```python
morph_lem = [morph.parse(token)[0][2] for token in stem_tokens]
morph_lem[:10]
```

```python
for n in range(len(stem_lem)):
    if stem_lem[n] != morph_lem[n]:
        print(stem_tokens[n], stem_lem[n], morph_lem[n])
```

* Mystem дает лемму несовершенного вида для глагольных токенов как совершенного, так и несовершенного вида (расскажу – рассказывать; отследить – отслеживать; начнет – начинать), в то время как Pymorphy – лемму совершенного вида (расскажу – рассказать; отследить – отследить; начнет – начать).
* Mystem приводит прилагательные сравнительной степени к наречиям (активнее – активно; больше – много; точнее – точно), а Pymorphy – либо к прилагательным начальной степени (активнее – активный; больше – большой), либо к прилагательным сравнительной степени (наибольшую – больший; точнее – точнее).
* Mystem правильно разбирает заимствования (юзабилиста – юзабилист; юзабилити – юзабилити), а Pymorphy – нет (юзабилиста – юзабилистый; юзабилити – юзабилить).
* Pymorphy, в отличие от Mystem, умеет расшифровывать сокращения (т.е. = так быть).
* Pymorphy различает е и ё.
* Анализаторы по-разному разбирают местоимения: для майстема *им* – творительный падеж местоимения *он* (этот меч выкован *им*), а для пайморфи *им* – дательный падеж местоимения *они* (дать *им* денег). Правильным является вариант пайморфи: "...В этом *им* поможет Сумасшедшее яйцо...".
* Кажется, что майстем чуть лучше по качеству в среднем, а пайморфи иногда чуть лучше разбирает сложные случаи, наподобие приведенного выше.
